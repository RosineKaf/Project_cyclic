```python
import os

# Renommer les fichiers uploadés

# L'erreur indique que 'train.csv' est introuvable.
# Il est probable que 'train.csv' ait déjà été renommé en 'clients.csv'.
# Vérifions si 'train.csv' existe avant de tenter de le renommer.
# Si 'clients.csv' existe déjà, cela signifie que le renommage a déjà eu lieu.
if os.path.exists('train.csv') and not os.path.exists('clients.csv'):
    os.rename('train.csv','clients.csv') # ancien nom, nouveau nom
    print("Fichier 'train.csv' renommé en 'clients.csv'.")
elif os.path.exists('clients.csv'):
    print("Le fichier 'clients.csv' existe déjà (probablement 'train.csv' a été renommé précédemment).")
else:
    print("Le fichier 'train.csv' est introuvable pour le renommage.")

# Si l'intention était de renommer 'test.csv' en 'clients_eval.csv', nous devrions modifier la deuxième ligne.
# Vérifions si 'test.csv' existe avant de tenter de le renommer.
if os.path.exists('test.csv') and not os.path.exists('clients_eval.csv'):
    os.rename('test.csv','clients_eval.csv')
    print("Fichier 'test.csv' renommé en 'clients_eval.csv'.")
elif os.path.exists('clients_eval.csv'):
    print("Le fichier 'clients_eval.csv' existe déjà (probablement 'test.csv' a été renommé précédemment).")
else:
    print("Le fichier 'test.csv' est introuvable pour le renommage en 'clients_eval.csv'.")
```
